- 샘플을 본 결과, 리뷰의 맨 처음엔 About this item이라는 말이 공통적으로 들어간다
- 이는 공통적으로 제거해준다

```python
df_1['contents_1']=df_1['내용'].str[16:]
```

```python
# cleaning data -> lower
df_1['lower_contents']=df_1['contents_1'].str.lower()
```

```python
# tokenization
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

all_words=[]

for sentence in df_1['lower_contents']:
  words=word_tokenize(sentence)
  for word in words:
    all_words.append(word)

print(all_words)
```

```python
# 불용어 제거

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))

filtered_words=[word for word in all_words if not str(word).lower() in stop_words]
```

```python
print(filtered_words)
```

```python
# lemmatization
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word) for word in filtered_words]
```

```python
print(lemmatized_words)
```

```python
# 특수문자 제거
import re
compile = re.compile("[^a-zA-Z]+")

final_words=[]

for i in range(len(lemmatized_words)):
    a = re.sub(compile, "", lemmatized_words[i])
    if a=="": continue
    final_words.append(a)
print(final_words)
```
